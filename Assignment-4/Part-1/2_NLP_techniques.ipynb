# NLP Techniques

# Extract the web URL text using BeautifulSoup
## Save the result or extracted paragraphs in a file “input.txt”

```python
# import libraries
from bs4 import BeautifulSoup
import requests

# create objects for the BeautifulSoup library
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
webpage = requests.get(url)
webpage_content = BeautifulSoup(webpage.content, "html.parser")

# extract first paragraph from the web page text
first_para = webpage_content.find_all('p')[0].text
print(first_para)

# Write the paragraph content in the 'input.txt' file for writing
# open the input file to write the contents
infile = open("input", "w", encoding='utf-8')
infile.write(first_para)
```

# Tokenization
#### Coverting sentences to words

```python
from nltk.tokenize import sent_tokenize, word_tokenize

# Read the input file text
infile = open('input', 'r', encoding='utf-8')
text = infile.read()
```

#### Sentence Tokenization

```python
# Printing all the sentences
s_tokens = sent_tokenize(text)
for i in s_tokens:
    print(i)
```

#### Word Tokenization

```python
# Printing all the words
w_tokens = word_tokenize(text)
for i in w_tokens:
    print(i)
```

# POS
#### Part of Speech tagging

```python
import nltk
nltk.download('averaged_perceptron_tagger')

# Pass the word tokens processed previously to get Part of Speech tagging output
pos_tagged = nltk.pos_tag(w_tokens)

print("Part of Speech tagging output: ", pos_tagged)
```

# Stemming
#### Words are reduced to a root by removing inflection through dropping unnecessary characters, usually a suffix

```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

pStemmer = PorterStemmer()
lStemmer= LancasterStemmer()
sStemmer = SnowballStemmer('english')
w_tokens = word_tokenize(text)
print("PorterStemmer output :")
for w in w_tokens:
    print(pStemmer.stem(w))
    
print("\nLancasterStemmer output :")
for w in w_tokens:
    print(lStemmer.stem(w))
    
print("\nSnowballStemmer output :")
for w in w_tokens:
    print(sStemmer.stem(w))
    
```

# Lemmatization
#### Another approach to remove inflection by determining the part of speech and utilizing detailed database of the language

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
print("Lemmatization output: ")
for w in w_tokens:
    print(lemmatizer.lemmatize(w))
```

# Trigram
#### A set of co-occurring or continuous sequence of n items from a sequence of large text or sentence

```python
import nltk
from nltk import word_tokenize

print("Trigram output: \n")

for s in s_tokens:
    wtokens = nltk.word_tokenize(s)
    for i in range(len(wtokens) - 2):
        print(wtokens[i], wtokens[i + 1], wtokens[i + 2])
```

# Named Entity Recognition
#### Subtask of information extraction that seeks to locate and classify elements in text into pre-defined categories

```python
import nltk
from nltk import wordpunct_tokenize, pos_tag, ne_chunk

nltk.download('maxent_ne_chunker')
nltk.download('words')

for s in s_tokens:
    x= ne_chunk(pos_tag(wordpunct_tokenize(s)))
    print(x)
```

# Plot the Frequencies of words

```python
from collections import Counter
from nltk.tokenize import RegexpTokenizer

# Use tokenizer to remove all the regular expression and extract only words
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text.lower())

# Create a non-empty dictionary
word_count = {}

# Start storing the words and their occurances in the dictionary
for word in tokens:
    if word in word_count.keys():
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print the contents of dictionary
for x, y in word_count.items():
    print(x + " : ", y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(15,5))
sns.set_theme(style="darkgrid")

data_words = word_count.keys()
words_counts = word_count.values()
indexes = np.arange(len(data_words))
width = 0.7

plt.title('Plot for Frequencies of words', size=14)
plt.xlabel('word', size=14)
plt.xticks(size=10, rotation= 90)
plt.ylabel('frequency', size=14)
plt.yticks(size=10)
plt.bar(indexes, words_counts, width)
plt.xticks(indexes + width * 0.5, data_words)
plt.show()
```
